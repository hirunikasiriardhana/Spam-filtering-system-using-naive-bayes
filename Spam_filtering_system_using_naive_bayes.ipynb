```python
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()#to upload data set file
```

```python
import io# to get file
df = pd.read_csv(io.StringIO(uploaded["spam.csv"].decode('ISO-8859-1')))
```

```python
df.head()
```

```python
#Add two columns as label(v1) and message(v2)
df['label'] = df['v1'].astype('str')
df['message'] = df['v2'].astype('str')
```

```python
#Remove other columns
df.drop(labels=['Unnamed: 2','Unnamed: 3', 'Unnamed: 4', 'v1', 'v2'],axis=1, inplace=True)
```

```python
df.head()
```

```python
#Dividing to training and testing set
from sklearn.model_selection import  train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.message,df.label,test_size=0.3)
```

```python
#Convert msg to numeric format
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
```

```python
#Convert msg to vector format (add both train and test data set )
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
#there are different classifier metods binomial and multinomial.we use multinomial method because this problem have multiple
#features(in message column)
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
#Train the model
model.fit(X_train,y_train)
```

```python
#Get score
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test[0:1])
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
predictions = model.predict(X_test)
print('Accuracy: ', accuracy_score(y_test,predictions))
print('Precision: ', precision_score(y_test,predictions,pos_label='ham'))
print('Recall: ', recall_score(y_test,predictions,pos_label='ham'))
```

```python
def classify_spam(arr):
  return model.predict(vectorizer.transform(arr))
```
